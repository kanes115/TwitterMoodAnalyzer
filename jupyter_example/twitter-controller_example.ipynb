```python
import twitter_controller
```

```python
import twitter_controller.Person
```

```python
# We can create entities which represents people
donald_trump = twitter_controller.Person('Donald Trump')
```

```python
# Let's take a look at his face, here's URL
donald_trump.get_profile_pict()
```

```python
# But let's get to the point - TWEETS (as Text objects)
donald_trump.get_texts()
```

```python
# What is Text?
import twitter_controller.Text
t = twitter_controller.Text('Here goes content', hashtags=['MakeAmericaAWall', 'LikeInChina'], links=['http://www.bbc.com/donald_trump'])
```

```python
print(t)
```

```python
# You can add texts, hashtags are not being 
p = twitter_controller.Text('Tralalala', hashtags=['Powder', 'LikeInChina'], links=['http://www.justanotherlink.com'])
print(t + p)
```
